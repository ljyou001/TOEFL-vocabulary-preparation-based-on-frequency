# dealing with Text

```python
def split_into_words(line):
    import re
    word_regex_improved = r"(\w[\w']*\w|\w)"
    word_matcher = re.compile(word_regex_improved)
    return word_matcher.findall(line)
```

```python
import nltk
# nltk.download('punkt')
# nltk.download('stopwords')
```

```python
with open('./text/all_in_one.txt', 'r', encoding='utf-8') as f:
    s = f.read()
s = s.lower()
words = nltk.word_tokenize(s)
```

```python
words_cleaned = [i for i in words if len(i) > 1]
words_cleaned = [i for i in words_cleaned if i not in nltk.corpus.stopwords.words('english')]
```

```python
with open('./text/cleaned_words.txt', 'w', encoding='utf-8') as f:
    f.write(' || '.join(words_cleaned))
```

```python
from collections import Counter
counted = Counter(words_cleaned)
count_sorted = sorted(counted.items(), key=lambda x: x[1], reverse=True)
```
